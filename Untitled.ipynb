```python
radii = np.arange(1, 14, 1)   # change range if needed
fluxes = []



for rAperture in radii:
    
    rSkyInner = rAperture * 1.5
    rSkyOuter = rAperture * 2.0

    aperture = CircularAperture((xcor, ycor), r=rAperture)
    annulus = CircularAnnulus((xcor, ycor), r_in=rSkyInner, r_out=rSkyOuter)
    
    
    sky_table = aperture_photometry(WASP_image1, annulus)
    area = np.pi * (rSkyOuter**2 - rSkyInner**2)
    skyFlux = sky_table['aperture_sum'].value
    meanSky = skyFlux / area

    
    phot_table = aperture_photometry(WASP_image1 - meanSky, aperture)
    flux = phot_table['aperture_sum'].value
    fluxes.append(flux)


plt.figure(figsize=(7,5))
plt.plot(radii, fluxes, marker='o')
plt.xlabel('Aperture radius (pixels)', fontsize=12)
plt.ylabel('Sky-subtracted counts', fontsize=12)
plt.title('Curve of Growth', fontsize=14)
plt.grid(True)
plt.show()

# curve of growth varies ar higher r as alot of stars around wasp
```
